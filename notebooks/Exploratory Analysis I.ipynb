```python
import glob
import pandas as pd
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
KNOWN_IMAGES = "../data/dataset1/known_images/"
```

```python
primitive_hash = list()
```

```python
#read all filepaths in dataset1
for file_path in glob.glob(KNOWN_IMAGES+"*.png"):
    file_name = os.path.basename(file_path)
    category = file_name.split("_")[0]
    primitive_hash.append({"category":category,"im_path":file_name})
```

```python
df = pd.DataFrame(primitive_hash)
```

```python
df.head(5)
```

```python
df['category'].value_counts().plot(kind="bar")
```

```python
class_names = list()
images = list()
for k in df.sample(6).itertuples():
    class_names.append(k[1])
    img = cv2.imread(KNOWN_IMAGES+k[2])
    images.append(img)
```

```python
plt.figure(figsize=(10,10))
columns = 3
for i in range(6):
    plt.subplot(int(len(images)/ columns + 1), columns, i + 1)
    plt.imshow(images[i])
    plt.title('Image #{} : '.format(i+1) + class_names[i])
```
